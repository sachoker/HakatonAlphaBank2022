```python
import pandas as pd
import numpy as np
from shapely.geometry import Point
from shapely.geometry.polygon import Polygon
```

```python
raw_data = pd.read_csv('data/train/some_data_4.csv')
raw_data = raw_data.loc[:, ~raw_data.columns.str.contains('^Unnamed')]
raw_data.drop(columns=['lat_h3', 'lon_h3'], inplace=True)
raw_data.dropna(inplace=True)
mall_data = pd.read_csv('data/train/mall_coord.csv')
mall_data = mall_data.loc[:, ~mall_data.columns.str.contains('^Unnamed')]
raw_data
```

```python
raw_data.insert(3, 'mall', 0)
raw_data
```

```python
for i, row in raw_data.iterrows():
    poly = row['walk_15min'][10:-2].split(',')
    for n, j in enumerate(poly):
        j = j.strip()
        poly[n] = list(map(float, j.split(' ')))
    for j, mall in mall_data.iterrows():
        lat=mall['mall_lat']
        lng=mall['mall_lng']
        pnt = Point(lng, lat)
        polygon = Polygon(poly)
        if polygon.contains(pnt):
            raw_data.loc[raw_data.index==i, 'mall'] = 1
            break
            break
```

```python
sum_data = pd.read_csv('data/train/some_data.csv')
sum_data.tail()
```

```python
raw_data.drop(columns=['geo_h3_10', 'avg', 'walk_15min'], inplace = True)
raw_data
```
